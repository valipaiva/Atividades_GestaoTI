```python
#Realizar o upload do arquivo STORES.csv.
#Renomear todas as colunas do arquivo STORES.csv, onde os respectivos nomes sejam compactados (Ex: Daily_Customer_Count foi renomeado para Visitantes).
#Após isto, para se analisar o desempenho das lojas de supermercado/mercado do arquivo STORES.csv encontre os valores mínimo, máximo, médio e
#desvio padrão das seguinte colunas: "Items_Available"; "Daily_Customer_Count"; e "Store_Sales".
#Posto isto, realizar três gráficos com as seguintes informações: Items_Available, Daily_Customer_Count e Store_Sales. #
#Não se esqueça de colocar: nomes para os eixos Y e X do gráfico, colocar cores diferentes para os pontos e colocar legenda.


from google.colab import files
import pandas as pd

# Fazer o upload do arquivo
uploaded = files.upload()

# Carregar o arquivo CSV
df = pd.read_csv(next(iter(uploaded)))
df.head()

# Renomeando as colunas
df.rename(columns={
    'Daily_Customer_Count': 'Visitantes',
    'Items_Available': 'Itens_Disponiveis',
    'Store_Sales': 'Vendas_Loja'
}, inplace=True)

df.head()
# Calcular estatísticas para as colunas
estatisticas = {
    'Itens_Disponiveis': {
        'Min': df['Itens_Disponiveis'].min(),
        'Max': df['Itens_Disponiveis'].max(),
        'Média': df['Itens_Disponiveis'].mean(),
        'Desvio Padrão': df['Itens_Disponiveis'].std()
    },
    'Visitantes': {
        'Min': df['Visitantes'].min(),
        'Max': df['Visitantes'].max(),
        'Média': df['Visitantes'].mean(),
        'Desvio Padrão': df['Visitantes'].std()
    },
    'Vendas_Loja': {
        'Min': df['Vendas_Loja'].min(),
        'Max': df['Vendas_Loja'].max(),
        'Média': df['Vendas_Loja'].mean(),
        'Desvio Padrão': df['Vendas_Loja'].std()
    }
}

# Exibir as estatísticas
estatisticas
import matplotlib.pyplot as plt
import seaborn as sns

# Definir o estilo do gráfico
sns.set(style="whitegrid")

# Gráfico para "Itens_Disponiveis"
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['Itens_Disponiveis'], c='blue', label='Itens Disponíveis', alpha=0.6)
plt.title('Itens Disponíveis nas Lojas', fontsize=16)
plt.xlabel('Índice', fontsize=12)
plt.ylabel('Itens Disponíveis', fontsize=12)
plt.legend()
plt.show()

# Gráfico para "Visitantes"
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['Visitantes'], c='green', label='Visitantes', alpha=0.6)
plt.title('Visitantes Diários nas Lojas', fontsize=16)
plt.xlabel('Índice', fontsize=12)
plt.ylabel('Visitantes', fontsize=12)
plt.legend()
plt.show()

# Gráfico para "Vendas_Loja"
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['Vendas_Loja'], c='red', label='Vendas na Loja', alpha=0.6)
plt.title('Vendas nas Lojas', fontsize=16)
plt.xlabel('Índice', fontsize=12)
plt.ylabel('Vendas', fontsize=12)
plt.legend()
plt.show()
```
